```python
import pandas as pd

data = {
    'observed_field': [
        'first name', 'fname', 'given name', 
        'last name', 'lname', 'surname', 
        'email address', 'mail id', 'contact email', 
        'phone number', 'cell', 'mobile no', 'contact no', 
        'expected ctc', 'desired salary', 'expected compensation', 'salary expectations',
        'current ctc', 'ctc expected', 'what is your ctc', # <- Added more CTC variations
        'current location', 'city of residence', 'where do you live', 'where do you reside', 'country of residence',
        'linkedin url', 'linkedin profile', 'link to linkedin'
    ],
    'master_key': [
        'first_name', 'first_name', 'first_name', 
        'last_name', 'last_name', 'last_name', 
        'email', 'email', 'email', 
        'phone', 'phone', 'phone', 'phone', 
        'expected_salary', 'expected_salary', 'expected_salary', 'expected_salary',
        'expected_salary', 'expected_salary', 'expected_salary',
        'location', 'location', 'location','location', 'location',
        'linkedin', 'linkedin', 'linkedin'
    ]
}

df = pd.DataFrame(data)
print("Training data loaded! Total samples:", len(df))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
import joblib

# 1. Create pipeline WITH stop_words to ignore "what", "is", "your", etc.
model = make_pipeline(
    TfidfVectorizer(ngram_range=(1, 2), stop_words='english'), 
    RandomForestClassifier(n_estimators=100, random_state=42)
)

# 2. Train the model
model.fit(df['observed_field'], df['master_key'])
print("✅ Model trained successfully!")

# 3. Test the model's intelligence
test_fields = ["what is your ctc", "mobile phone", "linkedin link", "where do you reside"]
predictions = model.predict(test_fields)

for field, pred in zip(test_fields, predictions):
    print(f"I see '{field}' -> I predict it means: '{pred}'")
```
